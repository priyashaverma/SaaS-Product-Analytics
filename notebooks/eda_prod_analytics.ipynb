```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("saas_customer_data.csv", parse_dates=['signup_date'])


churn_rate = df['churned'].mean()
print(f"Churn Rate: {churn_rate:.2%}")


sns.countplot(x='plan', data=df)
plt.title("Plan Distribution")
plt.show()


revenue_by_plan = df.groupby('plan')['monthly_revenue'].sum().sort_values()
revenue_by_plan.plot(kind='barh', color='skyblue')
plt.title("Revenue by Plan")
plt.xlabel("Total Monthly Revenue")
plt.show()


sns.boxplot(x='churned', y='sessions_last_month', data=df)
plt.title("Sessions vs Churn")
plt.show()


df['signup_month'] = df['signup_date'].dt.to_period('M')
signup_growth = df.groupby('signup_month')['user_id'].count()
signup_growth.plot(figsize=(12,6))
plt.title("User Signup Growth Over Time")
plt.ylabel("New Users")
plt.show()
```
