```python
import pandas as pd
import numpy as np


df = pd.read_csv("saas_customer_data.csv", parse_dates=['signup_date'])

np.random.seed(42)
df['group'] = np.random.choice(['A', 'B'], size=len(df), p=[0.5, 0.5])


df['sessions_last_month_ab'] = df.apply(lambda row: 
    row['sessions_last_month'] + (np.random.randint(1,4) if row['group']=='B' else 0), axis=1)

df['churned_ab'] = df.apply(lambda row: 
    row['churned'] if row['group']=='A' else np.random.choice([0,1], p=[0.85,0.15]), axis=1)


df.head()
```
